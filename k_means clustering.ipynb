```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(dfincome)
    wc=km.inertia_
    wcss.append(wc)
print(wcss)    
```

```python
plt.plot(range(1,11),wcss)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(dfincome)
```

```python
df=ss.transform(dfincome)
```

```python
df
```

```python
wcss1=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(df)
    wc=km.inertia_
    wcss1.append(wc)
print(wcss1)
    
```

```python
plt.plot(range(1,11),wcss1)
plt.show()
```
